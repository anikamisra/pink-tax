Now let's clean the data. 

```python
w_df['Product Title'] = w_df['Product Title'].astype(str)
w_df['Product Title'] = w_df['Product Title'].str.replace('Nike', '', regex=False)
w_df['Product Title'] = w_df['Product Title'].str.replace('(', '', regex=False)
w_df['Product Title'] = w_df['Product Title'].str.replace(')', '', regex=False)

w_df['Product Subtitle'] = w_df['Product Subtitle'].astype(str)
w_df['Product Subtitle'] = w_df['Product Subtitle'].str.replace('Nike', '')
w_df['Product Subtitle'] = w_df['Product Subtitle'].str.replace('(', '')
w_df['Product Subtitle'] = w_df['Product Subtitle'].str.replace(')', '')

# remove apostrophes, they are causing issues
w_df['Product Title'] = w_df['Product Title'].str.replace("'", "", regex=False)
w_df['Product Subtitle'] = w_df['Product Subtitle'].str.replace("'", "", regex=False)


# ensure clothes are specifically labelled "womens"  
w_df = w_df[~w_df['Product Title'].str.contains("Mens")]
w_df = w_df[~w_df['Product Title'].str.contains(" mens")]
w_df = w_df[~w_df['Product Subtitle'].str.contains("Mens")]
w_df = w_df[~w_df['Product Subtitle'].str.contains(" mens")]
w_df = w_df[
    w_df['Product Title'].str.lower().str.contains("womens", na=False) |
    w_df['Product Subtitle'].str.lower().str.contains("womens", na=False)
]

w_df.head()
```

```python
m_df['Product Title'] = m_df['Product Title'].astype(str)
m_df['Product Title'] = m_df['Product Title'].str.replace('Nike', '', regex=False)
m_df['Product Title'] = m_df['Product Title'].str.replace('(', '', regex=False)
m_df['Product Title'] = m_df['Product Title'].str.replace(')', '', regex=False)

m_df['Product Subtitle'] = m_df['Product Subtitle'].astype(str)
m_df['Product Subtitle'] = m_df['Product Subtitle'].str.replace('Nike', '')
m_df['Product Subtitle'] = m_df['Product Subtitle'].str.replace('(', '')
m_df['Product Subtitle'] = m_df['Product Subtitle'].str.replace(')', '')

# remove apostrophes, they are causing issues
m_df['Product Title'] = m_df['Product Title'].str.replace("'", "", regex=False)
m_df['Product Subtitle'] = m_df['Product Subtitle'].str.replace("'", "", regex=False)


# ensure clothes are specifically labelled "mens"  
m_df = m_df[~m_df['Product Title'].str.lower().str.contains("womens")]
#m_df = m_df[~m_df['Product Title'].str.contains("womens")]
m_df = m_df[~m_df['Product Subtitle'].str.lower().str.contains("womens")]
#m_df = m_df[~m_df['Product Subtitle'].str.contains("womens")]
m_df = m_df[
    m_df['Product Title'].str.lower().str.contains("mens", na=False) |
    m_df['Product Subtitle'].str.lower().str.contains("mens", na=False)
]

m_df.tail()
```

```python
# see how many unique words are in the subtitle (we will just start with subtitle)
all_words_list = []
for item in w_df['Product Subtitle']: 
    if (isinstance(item, str) == True) :
        item_list = item.split()
        for word in item_list: 
            word = word.lower()
            all_words_list.append(word)
for item in m_df['Product Subtitle']: 
    if (isinstance(item, str) == True) : 
        item_list = item.split()
        for word in item_list: 
            word = word.lower()
            all_words_list.append(word)

all_words_list = sorted(all_words_list)
all_words_set = set(all_words_list)
print(all_words_set)
```

Now we will make a feature matrix. The rows will be the samples (aka the products), the column features will be each unique word, and the value of each entry will be either 1 or 0 : whether or not that word appears in the product description. 

The output variable will be the price. 

Essentially this is a feature selection problem. We can then split the data into 80-20 to test the accuracy of each different feature selection model. 

Maybe I'll start with just the womens for now! 

```python
all_data = pd.DataFrame(columns=list(all_words_set))
all_data['Price'] = 0
all_data.head()

for index, row in w_df.iterrows(): 
    product = row['Product Subtitle']
    price = row['Price']

    newrow = {word: 0 for word in list(all_words_set)}
    newrow['Price'] = price

    words = set(product.lower().split())
    for word in words: 
        if word in newrow: 
            newrow[word] = 1 
    newrow['womens'] = 1

    all_data.loc[index] = newrow

all_data.head()


```
