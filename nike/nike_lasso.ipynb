```python
# lasso time 

x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
x_train.to_csv("x_train.csv", index=False)
x_test.to_csv("x_test.csv", index=False)
y_train.to_csv("y_train.csv", index=False)
y_test.to_csv("y_test.csv", index=False)
```

```python
x_train = pd.read_csv("x_train.csv")
x_train.shape
```

```python
y_train = pd.read_csv("y_train.csv")
y_train.shape
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=1)
```

```python
lasso.fit(x_train, y_train)
```

```python
y_pred = lasso.predict(x_test)
```

```python
y_pred
```

```python
print("Slope: %.10f" % lasso.coef_[0])
```

```python
print("Intercept: %.2f" % lasso.intercept_)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

```python
from scipy.stats import pearsonr
corr, _ = pearsonr(y_test, y_pred)
print("Pearson correlation coefficient:", corr)
```

```python
coefs = lasso.coef_
feature_names = x_train.columns
coefs_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': coefs})
coefs_df.to_csv("coefs.csv", index=False)
```

```python
elite = df[df['elite'] == 1]
no_elite = df[df['elite'] == 0]
```

```python
elite.describe()
```

```python
no_elite.describe()
```

```python
# repeat everything on no_elite. 

from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso

x_ne = no_elite.iloc[:, :-1]
y_ne = no_elite.iloc[:, -1]
x_ne_train, x_ne_test, y_ne_train, y_ne_test = train_test_split(x_ne, y_ne, random_state=0)
lasso_ne = Lasso(alpha=1)
lasso_ne.fit(x_ne, y_ne)
y_ne_pred = lasso_ne.predict(x_ne_test)
```

```python
from scipy.stats import pearsonr
corr, _ = pearsonr(y_ne_test, y_ne_pred)
print("Pearson correlation coefficient:", corr)
```

```python
coefs_ne = lasso_ne.coef_
feature_names_ne = x_ne_train.columns
coefs_ne_df = pd.DataFrame({'Feature': feature_names_ne, 'Coefficient': coefs_ne})
coefs_ne_df.to_csv("coefs_ne.csv", index=False)
```

Features with the most infleunce on price 


```python
coefs_1 = pd.read_csv("coefs.csv")
coefs_1.head()
```

```python
coefs_2 = pd.read_csv("coefs_ne.csv")
coefs_2.head()
```

```python
coefs_1 = coefs_1[coefs_1['Coefficient'] != 0.0]
print("These are the words that are most influential on price. \n")
print(coefs_1)
```

```python
coefs_2 = coefs_2[coefs_2['Coefficient'] != 0.0]
print("Since 'elite' had such a huge coefficent, these are the results \nfor when we take out all products that have the term 'elite' in their description. \n")
print(coefs_2)
```

In both cases, we can see that the coefficient for womens is negative. This implies that as "womens" decreases, price increases. But womens is a variable that has 0 = men, and 1 = women. So actually, that means that the model expects a word like "men" to INCREASE the price. That is even when we take out "elite" items. Pretty interesting!
